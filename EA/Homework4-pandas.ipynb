## Task 1 - DataFrame manipulation  

Using **states_covid.csv**, we are going to read the data in as a DataFrame to manipulate, subset, and filter in various ways.   

### Task 1.1 

Read in states_covid.csv with date as a "date" dtype, and only columns consisting of the hospitalization (4 col), ICU (2 col), and Ventilators (2 col)  

```python
covid_DF = pd.read_csv('states_covid.csv', usecols=[0,1,6,7,8,9,10,11,17,18], parse_dates=[1], infer_datetime_format=True)
covid_DF.head()
```

### Task 1.2 

For each of the following catergories: *currently* hospitalized, *currently* in the ICU, and *currently* on ventilation...  
Find the 5 states with the greatest numbers in each catergory and list them in order.      
*hint*: sort_values, unique  

```python
sorted_DF = covid_DF.sort_values(by = ['hospitalized', 'inIcuCurrently', 'onVentilatorCurrently'], ascending = [False, False, False])

sorted_DF.head()
```

```python
critical_DF = covid_DF.iloc[:, [0,1,4,7,9]]
hospitalized_DF = critical_DF.sort_values('hospitalizedCurrently', ascending = False)
hos = hospitalized_DF.drop_duplicates(subset=['state'])
hos.head(5)
```

```python
ICU_DF = critical_DF.sort_values('inIcuCurrently', ascending = False)
icu = ICU_DF.drop_duplicates(subset=['state'])
icu.head(5)
```

```python
Vent_DF = critical_DF.sort_values('onVentilatorCurrently', ascending=False)
vent = Vent_DF.drop_duplicates(subset=['state'])
vent.head(5)
```

### Task 1.3 

Find the date in which each state crossed 1000 cumulative hospitilized covid patients

```python
criticaldate_DF = covid_DF.loc[:, ['date', 'state', 'hospitalizedCumulative']]
criticaldate_DF = criticaldate_DF.sort_values('date', ascending =True)
criticaldate_DF1= criticaldate_DF.dropna()
criticaldate_DF1['totalhospitalized'] = criticaldate_DF[['state', 'hospitalizedCumulative']].groupby('state')['hospitalizedCumulative'].cumsum()
criticaldate_DF1.groupby('state')['hospitalizedCumulative'].sum()
criticaldate_DF1.tail()
```

```python
criticaldate_DF2 = criticaldate_DF1.loc[:, ['date', 'state','totalhospitalized']]
criticaldate_DF2[criticaldate_DF2['totalhospitalized']>1000]
```
